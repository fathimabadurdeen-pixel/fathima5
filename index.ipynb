# Introduction to NumPy

NumPy is a powerful Python library used for fast and efficient numerical computing. It provides a special data structure called an **N-dimensional array (ndarray)**, often simply called an array.

Compared to regular Python lists, NumPy arrays are:

- Faster
- More memory-efficient
- Better suited for large datasets
- Designed for numerical and scientific computing

NumPy is foundational in Data Science, and **Pandas is built on top of NumPy**. Many of Pandas' core capabilities rely directly on NumPy arrays. Because of this, understanding NumPy is essential before diving deeper into data analysis.
