# Measures of Central Tendency  
## Introduction & The Mean

## From Data to Insights

Measures of **central tendency** summarize thousands of data points into a single meaningful number.  

When someone asks:

- “What’s our typical customer age?”
- “What’s our average sale value?”

They’re asking for a measure of central tendency.

Choosing the wrong measure can lead to misleading conclusions and poor business decisions.

###  What is Central Tendency?

Central tendency is a **single value** that represents the “center” or “typical” value of a dataset.

Instead of reviewing 10,000 data points, you summarize them into one meaningful number.


##  The Three Measures



###  Mean (Average)
The mathematical center — uses **all values**.

 Definition

The **mean** is calculated by:

Mean = (Sum of all values) ÷ (Number of values)

μ = (x₁ + x₂ + ... + xₙ) / n

#### Important Property

The mean is the **balance point** of the data.

It uses every single value — which makes it mathematically precise.

However…

 It is **sensitive to outliers** (extreme values).


#### When to Use the Mean

Use the mean when:

- Data is **symmetrical**
- There are **no extreme outliers**
- Data is **continuous numerical**
- You need precision for further calculations (variance, standard deviation)


#### When to Avoid the Mean

Avoid the mean when:

- Outliers are present
- Data is heavily skewed (e.g., income distribution)
- Working with categorical or ordinal data
- Extreme values are data errors

###  Median (Middle Value)
The physical center — 50% above, 50% below.


### The Median: The Middle Value

####  What is the Median?

The **median** is the middle value when data is sorted from smallest to largest.

It represents the **physical center** of the dataset:
- 50% of values are below it  
- 50% of values are above it  


####  How to Calculate the Median

 **Step 1: Sort the data**

 **Step 2: Find the middle value**

- **Odd number of values** → Median is the middle value  
- **Even number of values** → Median is the average of the two middle values  


#### Why the Median is Outlier-Resistant

The median only depends on **position**, not magnitude.

Extreme values do not drastically change it.


#### When to Use the Median

Use the median when:

- Outliers are present  
- Data is skewed  
- You want a value that represents most people  
- Working with ordinal data  
- Extreme values are legitimate but shouldn’t dominate  


#### When Not Ideal

Avoid median when:

- Data is perfectly symmetrical  
- You need further statistical calculations (variance requires mean)  
- Sample size is very small  
- You want mathematical precision


####  Mode (Most Frequent)
The most common value in the dataset.


 `The Mode: The Most Common Value`

##### What is the Mode?

The **mode** is the value that appears most frequently in a dataset.

Unlike mean and median, the mode:

- Works with **numbers**
- Works with **categories**
- Works with **text**
- Works with **any repeatable data**

 **Unique Advantage:**  
Mode is the **only** measure of central tendency that works with categorical (non-numeric) data.

##### When to Use the Mode

Use mode when:

- Working with **categorical data** (colors, brands, cities)
- Using **discrete values** (shoe sizes, ratings)
- Identifying the **most popular choice**
- Working with **nominal data** (no inherent order)
- You want what “most people” selected


##### When Mode Is Less Useful

Avoid mode when:

- Continuous data has no repeats
- All values are unique
- Mathematical calculations are needed
- Too many modes create ambiguity

#### Decision Framework

#### Decision Framework: Which Measure Should You Use?

Choosing the right measure is **systematic**, not random.

Your choice depends on:
1. Data type  
2. Presence of outliers  
3. Data distribution  
4. Business question  

---

#### The 3-Step Decision Tree

##### Step 1: What Type of Data?

##### → Categorical / Nominal (colors, brands, categories)
✔ **Use MODE** (only valid option)

##### → Numeric / Continuous (prices, ages, salaries)
➡ Continue to Step 2


##### Step 2: Are There Outliers or Skewness?

##### → Yes (extreme values, long tails)
✔ **Use MEDIAN**

##### → No (fairly symmetric distribution)
➡ Continue to Step 3


##### Step 3: Data is Clean and Symmetric

✔ **Use MEAN** (most precise)

### Standard Deviation & Variance
**Measures of Spread and Variability**

#### 1. Why Variability Matters
- Central tendency (mean, median, mode) shows typical value but hides **spread**.
- Two datasets can have the same mean but **very different experiences** (predictable vs unpredictable).
- Variability is crucial in:
  - Risk assessment
  - Quality control
  - Forecasting
  - Decision making
  - Detecting anomalies



#### Variance (σ²)
- Measures average **squared distance** from the mean.
- Formula: `σ² = Σ(xᵢ - μ)² / N`
- Squaring avoids cancellation of positive and negative deviations.
- **Limitation:** Units are squared → hard to interpret.


````python
# example

def sample_variance(data):
    n = len(data)
    mean = sum(data) / n
    return sum((x - mean) ** 2 for x in data) / (n - 1)
````

````python
sample_variance(df_sales["Unit_SellingPrice"])
````

````python
# example in pandas
df_sales["Unit_MakingCost"].var(ddof=0)
````

````python
# example in numpy
import numpy as np

var = np.var(df_sales["Unit_SellingPrice"])
var
````


#### Standard Deviation (σ)
- Square root of variance → back to original units.
- Formula: `σ = √Variance`
- Intuitive measure of spread around the mean.
- **68-95-99.7 Rule:** For normal distributions:
  - 68% within ±1 SD
  - 95% within ±2 SD
  - 99.7% within ±3 SD


#### 3. Sample vs Population

| Type | Formula | When to use |
|------|--------|-------------|
| Population SD (σ) | divide by N | All data available |
| Sample SD (s) | divide by N-1 | Subset of population (most real-world cases) |

> **Tip:** Use sample SD by default; N-1 corrects bias.



#### 5. Limitations
- Sensitive to **outliers** → consider IQR
- Hard to compare across **different scales** → use Coefficient of Variation
- Assumes **normal distribution** → check data shape

````python
# <!-- example -->
sd = np.sqrt(var)
sd
````

# Range and Interquartile Range (IQR)

## The Range: Simplest Spread Measure
- Difference between **maximum** and **minimum** values.  
- Formula: `Range = Maximum - Minimum`  
- **Limitation:** Sensitive to outliers.


````python
# example in pandas
range = df_sales["Unit_MakingCost"].max() - df_sales["Unit_MakingCost"].min()
range
````

````python
# example in numpy
np_range = np.max(df_sales["Unit_MakingCost"]) - np.min(df_sales["Unit_MakingCost"])
np_range
````

#### Understanding Quartiles
Quartiles divide data into four equal parts

| Quartile    | Percentile | Description       |
| ----------- | ---------- | ----------------- |
| Q1          | 25th       | 25% of data below |
| Q2 (Median) | 50th       | 50% of data below |
| Q3          | 75th       | 75% of data below |


#### interquartile Range (IQR): The Robust Alternative

Measures the middle 50% of data.

Formula: IQR = Q3 - Q1

Resistant to outliers


````python
# example in pandas
Q1 = df_sales["Unit_MakingCost"].quantile(0.25)
Q3 = df_sales["Unit_MakingCost"].quantile(0.75)

IQR = Q3 - Q1
IQR
````

````python
# example in numpy
Q1 = np.percentile(df_sales["Unit_MakingCost"], 25)
Q3 = np.percentile(df_sales["Unit_MakingCost"], 75)

IQR = Q3 - Q1
IQR
````

##### When to Use Range vs IQR

`Range`: Absolute min/max, simple, no outliers

`IQR`: Robust, handles outliers, compares datasets, detects anomalies

#### Detecting Outliers Using IQR

`Lower Bound = Q1 - 1.5 × IQR`

`Upper Bound = Q3 + 1.5 × IQR`


````python
# example
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

lower_bound, upper_bound
````

````python
# detect outliers
outliers = df_sales[
    (df_sales["Unit_MakingCost"] < lower_bound) |
    (df_sales["Unit_MakingCost"] > upper_bound)
]

outliers
````

````python
# remove outliers
df_no_outliers = df_sales[
    (df_sales["Unit_MakingCost"] >= lower_bound) &
    (df_sales["Unit_MakingCost"] <= upper_bound)
]
````

### Understanding Distribution Shapes

#### What is a Distribution?
- Shows how data values are spread.  
- Shape helps interpret variability and choose statistical methods.  
- Guides whether mean or median is more appropriate and detects outliers.


### Main Distribution Shapes

#### Normal (Symmetric)
- Bell curve, mean = median = mode  
- 68-95-99.7 rule applies  
- Values cluster around the mean  

````python
# example
# Set random seed for reproducibility
np.random.seed(42)

# Generate normal distribution data
data = np.random.normal(
    loc=50,      # mean
    scale=10,    # standard deviation
    size=1000    # number of samples
)
````

````python
# example
sns.displot(data=data ,kde=True)
````



#### Right-Skewed (Positive)
- Long right tail, mean > median > mode  
- Outliers pull mean higher  
- Examples: Income, house prices, website durations  
- Use **median** to represent central tendency  


````python
# example
data = np.random.lognormal(mean=2, sigma=0.5, size=1000)
sns.displot(data=data,kde=True)
````


#### Left-Skewed (Negative)
- Long left tail, mode > median > mean  
- Outliers pull mean lower  
- Examples: Easy test scores, age at death, product ratings  
- Use **median** to represent central tendency  


#### How to Identify Shape
1. **Compare mean & median**  
   - Mean ≈ Median → Symmetric  
   - Mean > Median → Right-skewed  
   - Mean < Median → Left-skewed

#### Coefficient of Variation (CV)


## Why CV Matters
- Standard deviation alone can’t compare variability across **different units or scales**.  
  - Example: Height (cm) vs Weight (kg), or $10 stock vs $1,000 stock.  
- CV expresses **SD as a percentage of the mean**, making comparisons fair.


CV = (Standard Deviation / Mean) × 100%

**Interpretation:**  
- CV = 25% → SD is 25% of the mean value.  


````python
# example
cv_percent = (df_sales["Unit_MakingCost"].std() /
              df_sales["Unit_MakingCost"].mean()) * 100

cv_percent
````

````python
# Function for multiple columns
def calculate_cv(data):
    return (data.std() / data.mean()) * 100

cv_results = df_sales[['Unit_MakingCost','Unit_MakingCost','Units_Sold']].apply(calculate_cv)
print(cv_results)
````

### Interpreting CV Values

| CV Range | Interpretation  | Example                        |
| -------- | --------------- | ------------------------------ |
| < 15%    | Low variability | Manufacturing, utility bills   |
| 15–30%   | Moderate        | Retail sales, body temperature |
| > 30%    | High            | Stock returns, weather, income |

#### Identifying Outliers: Statistical Methods
**Measures of Spread and Variability**  

#### What Are Outliers?
- Data points significantly different from the rest.  
- Can be **errors**, **natural extremes**, or **key insights** (e.g., fraud).  
- Affect statistics, models, and interpretations.

**Types of Outliers:**  

1. **Point Outliers (Univariate)** – Single unusual value  
   - Example: Salaries: $45K, $50K, $48K, $52K, $500K → $500K outlier  
2. **Contextual Outliers** – Unusual in a context  
   - Example: 30°C in winter vs summer  
3. **Collective Outliers** – Subset unusual together  
   - Example: Sudden spike in website traffic  


#### IQR Method (1.5 × IQR Rule)
- Uses Interquartile Range (Q3 − Q1) to find extreme values.  
- **Bounds:**  

````python
# example
Q1 = df_sales['Unit_MakingCost'].quantile(0.25)
Q3 = df_sales['Unit_MakingCost'].quantile(0.75)
IQR = Q3 - Q1
outliers = df_sales[(df_sales['Unit_MakingCost'] < Q1-1.5*IQR) | (df_sales['Unit_MakingCost'] > Q3+1.5*IQR)]
````

# Box Plots

````python
# Box Plot with All Important Parts Labeled

import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=100)
data = np.append(data, 120)  # Add an outlier

# Calculate box plot statistics
Q1 = np.percentile(data, 25)
Q2 = np.percentile(data, 50)  # Median
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Create box plot
plt.figure(figsize=(9, 6))
box = plt.boxplot(data, vert=True, patch_artist=False)

# Label key components
plt.text(1.05, Q1, "Q1 (25%)", verticalalignment='center')
plt.text(1.05, Q2, "Median (Q2)", verticalalignment='center')
plt.text(1.05, Q3, "Q3 (75%)", verticalalignment='center')
plt.text(1.05, lower_bound, "Lower Bound (1.5×IQR)", verticalalignment='center')
plt.text(1.05, upper_bound, "Upper Bound (1.5×IQR)", verticalalignment='center')

plt.title("Box Plot with All Important Parts Labeled")
plt.ylabel("Values")
plt.xticks([1], ["Sample Data"])

plt.show()

# Print calculated values for clarity
print("Q1:", Q1)
print("Median (Q2):", Q2)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
````
